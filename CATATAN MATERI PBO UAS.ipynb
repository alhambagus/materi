# inheritance
![](/foto/download.png)
Inheritance adalah kelas yang menggunakan kode yang terdapat di dalam kelas lain. Artinya, seorang anak dapat mewarisi tinggi atau warna mata orang tua. Anak tersebut juga bisa memiliki nama belakang yang sama dengan orang tua mereka.

konsep inheritance membuat sebuah struktur atau hierarchy class dalam bentuk kode program.class yang diturunkan bisa disebut sebagai class induk(parent class) sedangkan class yang menerima penurunan disebut child class

Kelas anak atau subclass dapat mewarisi metode dan variabel dari kelas induk atau kelas dasar.

# encapsulation

Encapsulation adalah proses pemaketan data bersama metode-metodenya dimana hal ini bermanfaat untuk menyembeunyikan rincian-rincian implementasi dari pemakai. Dalam sebuah objek yang mengandung variabel-variabel dan method-method, dapat ditentukan hak akses pada sebuah variabel atau method dari objek.
![](/foto/enc1.png)
![](/foto/enc2.png)

-class Robot(object):
   def __init__(self):
      self.a = 123
      self._b = 123
      self.__c = 123

obj = Robot()
print(obj.a)
print(obj._b)
print(obj.__c)

Garis bawah tunggal: private variabel, itu tidak boleh diakses secara langsung. Tetapi tidak berhenti disitu (kecuali Konvensi). Garis bawah ganda: variabel private, lebih sulit untuk akses tetapi masih mungkin.

Variabel private hanya dapat diubah dalam metode kelas dan tidak di luar kelas.
-class Car:
class Car:

    __maxspeed = 0
    __name = ""

    def __init__(self):
        self.__maxspeed = 200
        self.__name = "Supercar"

    def drive(self):
        print('driving. maxspeed ' + str(self.__maxspeed))

redcar = Car()
redcar.drive()
redcar.__maxspeed = 10  # will not change variable because its private
redcar.drive()
